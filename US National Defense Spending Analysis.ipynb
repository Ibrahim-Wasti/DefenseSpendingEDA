```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import missingno as msno
```

```python
df = pd.read_csv('National Defense Spending - national_defense_spending_and_investment_usafacts.csv')
df.head()
```

1. **Years**: Years from 1972 - 2020
2. **Total National Defense Spending and Investment (\\$):** Total amount spent 
3. **Total Investment (\\$):** Total amount invested 
4. **Total Expenditures (\\$):** Total amount spent on expenditures 
5. **Total Compensation of Personnel (\\$):** Total amount spent on compensation of personnel
6. **Military (\\$):** Sub-division of compensation of personnel. Total amount spent on military personnel 
7. **Civilian (\\$):** Sub-division of compensation of personnel. Total amount spent on civilian personnel 
8. **Total Consumption of capital and depreciation (\\$):** Total consumption of capital and depreciation 
9. **Total Durable goods purchased (\\$):** Total amount spent on durable goods 
10. **Aircraft (\\$):** Sub-division of durable goods purchased. Total amount spent on aircrafts 
11. **Missiles (\\$):** Sub-division of durable goods purchased. Total amount spent on missiles 
12. **Vehicles (\\$):** Sub-division of durable goods purchased. Total amount spent on vehicles
13. **Ships (\\$):** Sub-division of durable goods purchased. Total amount spent on ships 
14. **Electronics (\\$):** Sub-division of durable goods purchased. Total amount spent on electronics 
15. **Other (\\$):** Sub-division of durable goods purchased. Total amount spent on other durable goods 
16. **Total Nondurable goods purchased (\\$):** Total amount spent on nondurable goods 
17. **Petroleum products (\\$):** Sub-division of nondurable goods purchased. Total amount spent on petroleum products 
18. **Ammunition (\\$):** Sub-division of nondurable goods purchased. Total amount spent on ammunition 
19. **Other (\\$):** Sub-division of nondurable goods purchased. Total amount spent on other nondurable goods 
20. **Total Services purchased (\\$):** amount spent on services 
21. **Installation Support (\\\\$):** Sub-division of services purchased. Total amount spent on installation support
22. **Weapons Support (\\$):** Sub-division of services purchased. Total amount spent on weapons support 
23. **Personnel Support (\\$):** Sub-division of services purchased. Total amount spent on personnel support 
24. **Transportation of material (\\$):** Sub-division of services purchased. Total amount spent on transportation of material 
25. **Travel of persons (\\$):** Sub-division of services purchased. Total amount spent on travel of persons 
26. **Total Own-account investment and sales to other (\\$):** total invested in own-account and sales to other

```python
plt.style.use('ggplot')
plt.style.use('seaborn-talk-custom')
palette = sns.color_palette('muted')
palette
```

```python
sns.lineplot(data=df,x='Years',y=df.iloc[:,1],color=palette[3])
plt.title('Total Spending since 1972')
plt.ylabel('Total Spent in 100s of billions')
plt.xlabel('Years')
plt.xticks(ticks=np.arange(1972,2024,4))
plt.tight_layout()
```

- Total Spent on National Defense over the years

```python
sns.lineplot(data=df,x='Years',y=df.iloc[:,5],label='Military Personnel',color=palette[4])
sns.lineplot(data=df,x='Years',y=df.iloc[:,6],label='Civilian Personnel',color=palette[2])
plt.xticks(ticks=np.arange(1972,2024,4))
plt.title('Total Compensation (Military vs. Civilian)')
plt.ylabel('Compensation Paid ($)')
plt.tight_layout()
```

```python
sns.lineplot(data=df,x='Years',y=df.iloc[:,9],label='Aircraft',color=palette[1])
sns.lineplot(data=df,x='Years',y=df.iloc[:,10],label='Missiles',color=palette[2])
sns.lineplot(data=df,x='Years',y=df.iloc[:,11],label='Vehicles',color=palette[9],linestyle='--')
sns.lineplot(data=df,x='Years',y=df.iloc[:,12],label='Ships',color=palette[6],linestyle='--')
sns.lineplot(data=df,x='Years',y=df.iloc[:,13],label='Electronics',color=palette[7],linestyle='--')
sns.lineplot(data=df,x='Years',y=df.iloc[:,14],label='Other',color=palette[4])
plt.xticks(ticks=np.arange(1972,2024,4))
plt.title('Durable Goods Purchased by Subcategory')
plt.ylabel('Cost ($) in Billions')
plt.tight_layout()
```

```python
sns.lineplot(data=df,x='Years',y=df.iloc[:,16],label='Petroleum Products',color=palette[1])
sns.lineplot(data=df,x='Years',y=df.iloc[:,17],label='Ammunition',color=palette[2],linestyle='--')
sns.lineplot(data=df,x='Years',y=df.iloc[:,18],label='Other',color=palette[4],linestyle='--')
plt.xticks(ticks=np.arange(1972,2024,4))
plt.title('Nondurable Goods Purchased by Subcategory')
plt.ylabel('Cost ($) in Billions')
plt.tight_layout()
```
